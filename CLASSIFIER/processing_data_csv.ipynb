```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DATA_KAGGLE\sign_mnist_train.csv")
df.describe()
```

```python
df.head()
```

```python
#convert csv file to gray image
label_array = np.array(df.iloc[:,0])
label_1 = np.array(np.where(label_array == 15))
count = 0
for i in range(label_1.shape[1]):

    img = np.array(df.iloc[label_1[0][i],1:])
    img = np.uint8(img)
    img = np.reshape(img,(28,28))
    cv2.imwrite("DATA_KAGGLE\\UNKNOW\\" +"UNKNOW5_" + str(i) + ".jpg", img)
print("OK!!!")
```
